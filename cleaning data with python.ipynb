# Preparing Mexico Data

Import 

```python
df1 = pd.read_csv('/kaggle/input/dataset/mexico-real-estate-1.csv',  encoding='ISO-8859-1')
df2 = pd.read_csv('/kaggle/input/dataset/mexico-real-estate-1.csv',  encoding='ISO-8859-1')
df3 = pd.read_csv('/kaggle/input/dataset/mexico-real-estate-1.csv',  encoding='ISO-8859-1')
```

# 1.Clean df1

```python
df1=pd.read_csv('/kaggle/input/dataset/mexico-real-estate-1.csv',  encoding='ISO-8859-1',
               index_col=0)

print("df1 shape:",df1.shape)
print(df1.info())
df1.head()
```

remove NAN values

```python
df1.dropna(inplace=True)
df1.info()
df1.head()
```

Gettting rid of strings $ and transform price from object to float

```python
df1["price_usd"]=(
    df1["price_usd"]
    .str.replace("$","",regex=False)
    .str.replace(",","")
    .astype(float))
```

```python
df1["price_usd"].head()
```

# 2.Clean df2

```python
df2=pd.read_csv('/kaggle/input/dataset/mexico-real-estate-2.csv',  encoding='ISO-8859-1')
df2.drop(columns=["Unnamed: 0"],inplace=True)

print("df2 shape:",df2.shape)
print(df2.info())
```

```python
df2.head()
```

Remove 'NaN' values and create price_usd

```python
#Remove NaN values
df2.dropna(inplace=True)
#create price_usd
df2["price_usd"]=(df2["price_mxn"]/19).round(2)
df2.drop(columns=["price_mxn"],inplace=True)
```

```python
df2.head()
```

# 3.Clean df3

```python
df3=pd.read_csv('/kaggle/input/dataset/mexico-real-estate-3.csv',  encoding='ISO-8859-1')

print("df3 shape:",df3.shape)
print(df3.info())
```

```python
df3.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
df3.head()
```

Drop rows with NaN values in df3 and use the split method to create two new columns from "lat-lon" named "lat" and "lon"

```python
df3.dropna(inplace=True)
df3[["lat","lon"]]=df3["lat-lon"].str.split(",",expand=True)
df3.head()
```

Create state col from "place_with_parent_names,drop columns

```python
df3["state"]=df3["place_with_parent_names"].str.split("|",expand=True)[2]

#drop colums
df3.drop(columns=["place_with_parent_names","lat-lon"],inplace=True)
df3.head()
```

Concatenate DataFrames

```python
df =pd.concat([df1,df2,df3])
print(df.shape)
df.head()
```

```python
print(df.shape)
df.head()
```

# Save df

```python
df.to_csv('./mexico-real-estate-clean.csv',index=False)
```
